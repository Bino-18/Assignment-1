```python
# ML Assignment -1 
# Name: Berlin Bino, Bright Christel Pappa
# 700#: 700741472

import matplotlib.pyplot as plt
import numpy as np

# Create random vector of size 15 with integers in the range 1-20
random_vector = np.random.randint(1,21, size=15)

# Reshape the array to 3 by 5
reshaped_array = random_vector.reshape(3, 5)

# Print array shape
print("Array shape:", reshaped_array.shape)

# Replace the max in each row by 0
reshaped_array[np.arange(3), np.argmax(reshaped_array, axis=1)] = 0

# Print the modified array
print("Modified array:")
print(reshaped_array)

# Create a 2-dimensional array of size 4 x 3 with 4-byte integer elements
array_2d = np.zeros((4, 3), dtype=np.int32)

# Print shape, type, and data type of the array
print("Array shape:", array_2d.shape)
print("Array type:", type(array_2d))
print("Array data type:", array_2d.dtype)

# Define the square array
array = np.array([[3, -2], [1, 0]])

# Compute the eigenvalues and right eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(array)

# Print the eigenvalues
print("Eigenvalues:")
print(eigenvalues)

# Print the right eigenvectors
print("Right Eigenvectors:")
print(eigenvectors)

# Define the array
array = np.array([[0, 1, 2], [3, 4, 5]])

# Compute the sum of the diagonal elements
diagonal_sum = np.trace(array)

# Print the sum
print("Sum of diagonal elements:", diagonal_sum)

# Define the original array
original_array = np.array([[1, 2], [3, 4], [5, 6]])

# Reshape the array to 3x2
reshaped_array_3x2 = np.reshape(original_array, (3, 2))

# Print the reshaped array
print("Reshaped 3x2:")
print(reshaped_array_3x2)

# Reshape the array to 2x3
reshaped_array_2x3 = np.reshape(original_array, (2, 3))

# Print the reshaped array
print("Reshaped 2x3:")
print(reshaped_array_2x3)

# Define the programming languages and their popularity data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Find the index of the greatest popularity value
max_index = popularity.index(max(popularity))

# Create a list to store explosion values
explosion = [0.1 if i == max_index else 0 for i in range(len(languages))]

# Create a pie chart with exploded slice
plt.pie(popularity, labels=languages, autopct='%1.1f%%', explode=explosion)

# Set the chart title
plt.title('Popularity of Programming Languages')

# Display the chart
plt.show()
```
